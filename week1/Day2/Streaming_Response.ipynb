```python
from IPython.display import display, Markdown, update_display
```

```python
def yield_response():
    while True:
        yield 'Hello'
```

```python
markdown = display(Markdown("# Hello coming at a super sonic speed"), display_id = True)
generator = yield_response()
response = ""
```

```python
for i in range(500):
    response += next(generator)
    response += '\n'
    update_display(Markdown(response), display_id = markdown.display_id)
```

# Streaming response from Model

```python
from openai import OpenAI

OLLAMA_BASE_URI = "http://localhost:11434/v1"
OLLAMA_API_KEY = "CAN_BE_ANYTHING"
OLLAMA_MODEL = 'llama3.2'

client = OpenAI(
    base_url = OLLAMA_BASE_URI,
    api_key = OLLAMA_API_KEY
)
```

```python
message = [
    {"role": "user", "content": "Write a long essay on AI in 1000 words using markdown"}
]

stream = client.chat.completions.create(
    model = OLLAMA_MODEL,
    messages = message,
    stream = True
)

markdown = display(Markdown(""), display_id = True)
response = ""

for chunk in stream:
    response += ( chunk.choices[0].delta.content or '' )
    update_display(Markdown(response), display_id = markdown.display_id)
```
