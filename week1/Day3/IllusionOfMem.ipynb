```python
import gradio as gr
from litellm import completion

OLLAMA_MODEL = 'ollama_chat/llama3.2'
```

```python
SYSTEM_PROMT = """

    You are a helpful assistant in a clothes store. You should try to gently encourage \
    the customer to try items that are on sale. Hats are 60% off, and most other items are 50% off. \
    For example, if the customer says 'I'm looking to buy a hat', \
    you could reply something like, 'Wonderful - we have lots of hats - including several that are part of our sales event.'\
    Encourage the customer to buy hats if they are unsure what to get.

"""
```

```python
def chat(message, history):
    messsages = [{"role": "system", "content": SYSTEM_PROMT}]
    query = [{"role": "user", "content": message}]
    history = [{"role": h['role'], "content": h['content']} for h in history]
    messsages.extend(history)
    messsages.extend(query)
    
    stream = completion(model = OLLAMA_MODEL, messages = messsages, stream = True)

    response = ''
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(
    fn = chat, type = 'messages'
).launch()
```
