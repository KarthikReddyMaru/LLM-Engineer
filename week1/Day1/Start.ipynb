```python
# BASE URI and API KEY

from os import getenv
from dotenv import load_dotenv

load_dotenv(override=True)

OLLAMA_BASE_URI = 'http://localhost:11434/v1'
OLLAMA_API_KEY = 'CAN_BE_ANYTHING'
OLLAMA_MODEL = 'llama3.2'

GEMINI_BASE_URI = 'https://generativelanguage.googleapis.com/v1beta/openai/'
GEMINI_API_KEY = getenv('GEMINI_API_KEY')
GEMINI_MODEL = 'gemini-2.5-flash'
```

```python
# 1. API call to GEMINI
from openai import OpenAI

message = [
    {"role": "system", "content": "You are snarky AI assistant with a great humour"},
    {"role": "user", "content": "Who are you?"}
]

client = OpenAI(
    base_url = OLLAMA_BASE_URI,
    api_key = OLLAMA_API_KEY
)

response = client.chat.completions.create(
    model = OLLAMA_MODEL,
    messages = message
)

print(response.choices[0].message.content)
```

```python
# 2. Fetch and print available models
print("--- Supported Gemini Models (via OpenAI SDK) ---")
try:
    models = client.models.list()
    for model in models:
        # Google often prefixes models with "models/", e.g., "models/gemini-1.5-flash"
        print(f"ID: {model.id}") 
except Exception as e:
    print(f"Error listing models: {e}")
```
